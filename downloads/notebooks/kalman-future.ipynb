```python
import random
random.seed(12345)

import numpy as np
import matplotlib.pyplot as plt
from collections import namedtuple
```

Imagine you want to track the position of a ball which is being dropped off the Eiffel tower. To measure the ball's position we will setup a camera that can see the whole height of the tower and take many pictures. From the pictures we can then estimate the position of the ball at each of the moments that we took a picture.
The Eiffel tower is about 324m heigh. Let's say we have space for 20 pictures in our camera. It takes about 8.5s for the ball to reach the ground, so we will take a picture every 8.5/20 = 0.425seconds.

```python
starting_point = 324.
Nobservations = 20
time_step = 8.5/Nobservations

def plot_position(time, position, style='b-'):
    plt.plot(time, position, style)
    plt.xlabel("Time [s]")
    plt.ylabel("Height [m]")
```

This is the true position of the ball at each moment in time. It hits the ground around about 8s after we dropped it.

```python
true_positions = []
time = []
for n in range(Nobservations):
    t = n*time_step
    time.append(t)
    true_positions.append(starting_point - 0.5 * 9.81 * t**2)
    
plot_position(time, true_positions)
```

As the ball is dropping pretty quick it will probably be a bit blurry on our pictures. Plus to fit the whole Eiffel tower in the shot we will have to setup our camera quite a distance away. All this means when we measure the position we will invariably make a small mistake each time. Let's assume we can measure the position with an accuracy of about 10m. The uncertainty is pretty big, but that is on purpose. If we measured the position very accurately in the first place the effect of using the Kalman filter would be much less impressive. The point of this is to show that you can estimate the position of the ball accurately despite only measuring it with a large uncertainty at each time step. Below we plot the true position of the ball in blue and one possible observed path in red. Sometimes it looks like the ball went upwards!

```python
measurement_uncert = 10
# the first point we know, it is the top of the tower
observations = [starting_point]
for true_pos in true_positions[1:]:
    observations.append(true_pos + random.normalvariate(0, measurement_uncert))
    
plot_position(time, true_positions)
plot_position(time, observations, 'r')
```

But we can do better! By using what we know about the current position of the ball and Newton's laws we can predict where the next measurement should be. If you combine the prediction and the observation you get a better estimate of the true position then using either one by itself.
We could just take the average of the two as our combination, however you can do better than that by weighting the two by how certain you are each one is correct. If the measurement has a large uncertainty and the prediction has a small uncertainty we want to give a larger weight to the prediction, and vice versa. For this we need two ingredients:

* a method to predict the next position, and
* a method to estimate how certain we are that the prediction is correct

When the ball is let go at position $x$, $t$ seconds later it will be at $x'=x-\frac{1}{2}\cdot9.81\cdot t^2$. Once it is falling it will gather speed, $v'=v+9.81\cdot t$. As a result it will cover a larger distance during each time step: $x'=x-v\cdot t -\frac{1}{2}\cdot9.81\cdot t^2$.

```python
def kalman(x, P, measurement,
           R=np.matrix("10."),
           motion=np.matrix([-0.5*9.81*time_step**2, -9.81*time_step]).T,
           Q=np.matrix('''0. 0.;
                          0. 0.'''),
           F=np.matrix('''1. %f;
                          0. 1.'''%(time_step)),
           H=np.matrix('''1. 0.''')):
    """Parameters:
    x: initial state
    P: initial uncertainty convariance matrix
    measurement: observed position (same shape as H*x)
    R: measurement noise (same shape as H*x)
    motion: external motion added to state vector x
    Q: motion noise (same shape as P)
    F: next state function: x_prime = F*x
    H: measurement function: position = H*x

    Return: the updated and predicted new values for (x, P)
    """
    # Predict x and P
    x = F*x +motion
    P = F*P*F.T + Q
    print("xp:",x,"Pp:", P)
    
    # Update x, P using measurement m    
    # distance between measured and current position
    y = np.matrix(measurement).T - H * x
    S = H * P * H.T + R  # residual convariance
    K = P * H.T * S.I # Kalman gain or ratio
    print("P:",P)
    print("S.I:",S.I)
    print("H.T:", H.T)
    print("P*H.T:", P*H.T)
    print("H.T*S.I:", H.T*S.I)
    print("P*H.T*S.I:", P*H.T*S.I)
    print("X:",x)
    print("y:",y)
    x = x + K*y
    print("K:",K)
    print("K*y:",K*y)
    I = np.matrix(np.eye(F.shape[0])) # identity matrix
    P = (I - K*H)*P

    return x, P
```

```python
x = np.matrix('344. 0.').T
P = np.matrix(np.eye(2))*(20)
N = 1
predicted = []
for obs in observations[:N]:
    print("obs:",obs)
    print("x:",x,"P:",P)
    x,P = kalman(x,P,obs)
    print("x:",x,"P:",P)
    predicted.append(x.tolist()[0][0])
    
plot_position(time[:N], true_positions[:N])
plot_position(time[:N], observations[:N], 'ro')
plot_position(time[:N], np.array(observations[:N])+10, 'r--')
plot_position(time[:N], np.array(observations[:N])-10, 'r--')

plot_position(time[:N], predicted[:N], 'g')
```

```python
def next_position(x, v=0.):
    """Given we are at `x`, falling at speed `v` predict the next position and speed"""
    return x - v*time_step - 0.5 * 9.81 * time_step**2, v+9.81*time_step
def position_uncertainty(p, Q=0.5):
    """Estimate uncertainty of our prediction from the uncertainty on current position"""
    # current uncertainty plus a certain amount
    # which takes into account that our prediction
    # is not perfect, this depends strongly on
    # how good you think your extrapolation is
    return p + Q
```

```python
x = 324.; p = 2
print("Currently at: {:.2f}+-{:.3f}".format(x,p))
print("Prediction for next position: {:.2f}+-{:.3f}".format(next_position(x)[0],
                                                            position_uncertainty(p)))
```

```python
best_estimate_p = [starting_point+10]
best_estimate_v = [0.]
estimate_cov_p = [20]
estimate_cov_v = [20]

for n,(obs,t) in enumerate(zip(observations[1:], true_positions[1:])):
    next_p, next_v = next_position(best_estimate_p[n], best_estimate_v[n])
    p_cov = position_uncertainty(estimate_cov_p[n], Q=0)
    
    # this influences if we give more weight to our prediction
    # or more weight to the measurement
    ratio = p_cov/(p_cov + measurement_uncert**2)
    
    # best estimate of the position and uncertainty
    best_p = next_p + ratio*(obs - next_p)
    cov_p = (1-ratio) * p_cov
    best_v = next_v
    best_estimate_p.append(best_p)
    estimate_cov_p.append(cov_p)
    best_estimate_v.append(best_v)
    #estimate_uncert_v.append((1-ratio)*v_uncert)
    
```

```python
M=0;N=10
plot_position(time[M:N], true_positions[M:N])
plot_position(time[M:N], observations[M:N], 'ro')
plot_position(time[M:N], np.array(observations[M:N])+measurement_uncert, 'r--')
plot_position(time[M:N], np.array(observations[M:N])-measurement_uncert, 'r--')

plot_position(time[M:N], best_estimate_p[M:N], 'g')
plot_position(time[M:N],
              np.array(best_estimate_p[M:N])+np.sqrt(np.array(estimate_cov_p[M:N])), 'g--')
plot_position(time[M:N],
              np.array(best_estimate_p[M:N])-np.sqrt(np.array(estimate_cov_p[M:N])), 'g--')
```
