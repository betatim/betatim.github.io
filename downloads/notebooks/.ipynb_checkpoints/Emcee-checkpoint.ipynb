```python
import matplotlib.pyplot as plt

from matplotlib.ticker import MaxNLocator
```

```python
import numpy as np
import emcee
from scipy import special


ndim = 3
nwalkers = 250

def exgauss(x, mu, sig, tau):
    """Exponential convolved with gaussian resolution"""
    lam = 1./tau
    return (0.5*lam * np.exp(0.5*lam * (2*mu + lam*(sig**2) - 2*x)) *
            special.erfc((mu + lam*(sig**2) - x)/(np.sqrt(2)*sig)))
def gauss(x, mu, sig):
    return np.exp(-(((mu-x)**2)/(sig**2)))

# mixing a gauss and exp for the x dimension
def lnprob((x,y), mu, s, t):
    if x <= 50:
        return -np.inf
    return np.log((1*gauss(x, mu, s) + 5*np.exp(-(x/20.))) * exgauss(y, 0., 0.1, t))

def lnprob((x,y,z), mu_H, s_H, mu_W, s_W, t):
    return np.log(gauss(x, mu_H, s_H) * gauss(y, mu_W, s_W) * exgauss(z, 0., 0.1, t))

p0 = emcee.utils.sample_ball((80, 80., 0.), (2., 2., 1.), 250)

sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=[120., 2., 80.9, 4., 0.5])
```

```python
sampler.reset()
pos, prob, state = sampler.run_mcmc(p0, 400)
```

```python
fig, (ax_x, ax_y) = plt.subplots(2, 1, sharex=True, figsize=(8, 9))

ax_x.plot(sampler.chain[:, :, 0].T, color="k", alpha=0.4)
ax_x.yaxis.set_major_locator(MaxNLocator(5))
#ax_x.axhline(2., color="#888888", lw=2)
ax_x.set_ylabel("$x$")

ax_y.plot(sampler.chain[:, :, 1].T, color="k", alpha=0.4)
ax_y.yaxis.set_major_locator(MaxNLocator(5))
#ax_y.axhline(b_true, color="#888888", lw=2)
ax_y.set_ylabel("$y$")
```

```python
sampler.reset()
_ = sampler.run_mcmc(pos, 3000)
```

```python
for i in range(min(10, ndim)):
    plt.figure()
    plt.hist(sampler.flatchain[:,i], 100, color="k", histtype="step")
    plt.title("Dimension {0:d}".format(i))

plt.show()
```
